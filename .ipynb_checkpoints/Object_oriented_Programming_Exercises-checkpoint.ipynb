```python
#1.
class Calculator:
    def __init__ (self, number1, number2):
        self.number1 = number1
        self.number2 = number2
        
    def addition(self):
        return self.number1 + self.number2
    
    def subtraction(self):
        return self.number1 - self.number2
    
    def multiplication(self):
        return self.number1 * self.number2
    
    def division(self):
        try:
            return self.number1/self.number2
        except ZeroDivisionError:
            print("Can't divide it by zero.")
```

```python
c = Calculator(4,3)
print(c.division())
print(c.addition())
print(c.multiplication())
print(c.subtraction())
```

```python
#2.
class Book:
    def __init__(self, title, author, publication, year):
        self.title = title
        self.author = author
        self.publication = publication
        self.year = year
        
    def book_title(self):
        return self.title
    
    def book_author(self):
        return self.author
    
    def book_publication(self):
        return self.publication
    
    def book_year(self):
        return self.year
    
    def book_details(self):
        print("Title: {}".format(self.title))
        print("Author: {}".format(self.author))
        print("Publication: {}".format(self.publication))
        print("Year: {}".format(self.year))
```

```python
b = Book("Think and Grow Rich","Napoleon Hill","Imagine reads", 1952)
```

```python
print(b.book_details())
print("\n")
print(b.book_author())
print("\n")
print(b.book_publication())
print("\n")
print(b.book_year())
print("\n")
print(b.book_title())
```

```python
class Circle:
    def __init__(self,r):
        self.r = r
    def area(self):
        return 3.14 * self.r * self.r
    def perimeter(self):
        self.p = 2 * 3.14 * self.r
        return self.p
```

```python
c = Circle(3)
```

```python
c.area()
```

```python
c.perimeter()
```
